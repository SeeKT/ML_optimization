# Momentum

Artificial な関数に対して Momentum を用いる．

```python
########## Packages ##########
import autograd.numpy as np
from lib.lib import Test_function
from lib.plot import Plot_func
from lib.momentum import Momentum
##############################
```

```python
##### instance of the class #####
test = Test_function()
pf = Plot_func()
mmt = Momentum(maxiter=250, eps=1e-2, alpha=0.9)
#################################
```

```python
##### execute steepest descent #####
xinit = np.array([2.0, 1.5])
distort_mmt_x, itr_distort_mmt = mmt.momentum(
    func=test.distorted, xinit=xinit)
####################################
```

```python
x_range = np.linspace(-2.5, 2.5, 51); y_range = np.linspace(-2.5, 2.5, 51)
pf.plot_contour(
    func=test.sphere, x_range=x_range, y_range=y_range, 
    x_val=distort_mmt_x.T[0], y_val=distort_mmt_x.T[1], 
    itr=itr_distort_mmt, func_name="distorted", method_name="momentum")
```

```python
##### execute steepest descent #####
xinit = np.array([2.0, 2.5])
ackley_mmt_x, itr_ackley_mmt = mmt.momentum(
    func=test.ackley, xinit=xinit)
####################################
```

```python
x_range = np.linspace(-5, 5, 51); y_range = np.linspace(-5, 5, 51)
pf.plot_contour(
    func=test.ackley, x_range=x_range, y_range=y_range, 
    x_val=ackley_mmt_x.T[0], y_val=ackley_mmt_x.T[1], 
    itr=itr_ackley_mmt, func_name="ackley", method_name="momentum")
```

```python
##### execute steepest descent #####
xinit = np.array([2.0, 1.0])
matyas_mmt_x, itr_matyas_mmt = mmt.momentum(
    func=test.matyas, xinit=xinit)
####################################
```

```python
x_range = np.linspace(-0.5, 2.25, 51); y_range = np.linspace(-0.5, 2.25, 51)
pf.plot_contour(
    func=test.matyas, x_range=x_range, y_range=y_range, 
    x_val=matyas_mmt_x.T[0], y_val=matyas_mmt_x.T[1], 
    itr=itr_matyas_mmt, func_name="matyas", method_name="momentum")
```
